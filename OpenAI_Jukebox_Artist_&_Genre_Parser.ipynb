```python
import re


!wget https://raw.githubusercontent.com/deeplearn-art/jukebox/master/artist_genres.txt
file = "artist_genres.txt"

lines = []

with open(file) as fp:
   line = fp.readline()
   count = 1
   while line:
       print("Line {}: {}".format(count, line.strip()))
       line = fp.readline()
       lines.append(line)
       count += 1

split_string = []
artists = []
genres = []

for i in range(len(lines)):
    a_string = lines[i]
    split_string = a_string.split(":", 1)
    artists.append(split_string[0].replace("\'",""))
    

print(artists)

!wget https://github.com/openai/jukebox/blob/master/jukebox/data/ids/v3_genre_ids.txt
file1 = "v2_genre_ids.txt"

lines = []

with open(file1) as fp:
   line = fp.readline()
   count = 1
   while line:
       print("Line {}: {}".format(count, line.strip()))
       line = fp.readline()
       lines.append(line)
       count += 1

for i in range(len(lines)):
    a_string = lines[i]
    split_string = a_string.split(";", 1)
    genres.append(split_string[0])

print(genres)
```
